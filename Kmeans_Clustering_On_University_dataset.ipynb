```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Universities.csv")
```

```python
# Selecting features for clustering
features = df[['Top10']]

# Standardizing the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Trying different values of k
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)

# Plotting the elbow method
plt.figure(figsize=(8, 5))
sns.lineplot(x=range(1, 11), y=inertia, marker='o')
plt.title('Elbow Method to Find Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.show()
```

Elbow Graph Showing k must be 2.Therefore,2 clusters should be made

```python
df.corr()
```

```python
selected_features = df[['SAT', 'Top10', 'Accept', 'Expenses', 'GradRate']]


inertia_values = []
k_values = range(1, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(selected_features)
    inertia_values.append(kmeans.inertia_)

# Plot the elbow chart
plt.plot(k_values, inertia_values, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia (Sum of Squared Distances)')
plt.show()
```

By using Elbow method,we get the value of k=3

```python
# Selecting features for clustering
features = df[['SAT', 'Top10', 'Accept', 'Expenses', 'GradRate']]

# Standardizing the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Applying KMeans with k=2
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(scaled_features)

# Displaying the results
print(df[['Univ', 'Cluster']])
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
X=features
kmeans = KMeans(n_clusters=3)
labels = kmeans.fit_predict(X)

# Calculate Silhouette Score
silhouette_avg = silhouette_score(X, labels)
print(f"Silhouette Score: {silhouette_avg}")
```

```python
# Set the style for better visualization
sns.set(style="whitegrid")

# Scatter plot for 'Top10' and 'Accept' features
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Top10', y='Accept', hue='Cluster', data=df, palette='viridis', s=100)

# Adding labels and title
plt.title('KMeans Clustering of Universities')
plt.xlabel('Top10 Percentage')
plt.ylabel('Acceptance Rate')

# Display the plot
plt.show()
```

There seems to be a weak negative correlation between acceptance rate and the percentage of top 10 universities. This means that universities with higher acceptance rates tend to have a lower percentage of top 10 universities

```python
sns.set(style="whitegrid")

# Selecting features for clustering
features = df[['Top10']]

# Standardizing the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Applying KMeans with k=2
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(scaled_features)

# Scatter plot for 'Top10' feature
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Top10', y='Expenses', hue='Cluster', data=df, palette='viridis', s=100, legend='full')

# Adding labels and title
plt.title('KMeans Clustering of Universities based on Top10')
plt.xlabel('University')
plt.ylabel('Expenses')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.show()
```

This means that universities with higher average Expenses tend to have a higher percentage of students from the top 10%, and vice versa.

The data within each cluster appears to have some spread. This suggests that there might be further sub-groups or variations within each cluster
